```python
import nltk
nltk.download('punkt')
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from sklearn.naive_bayes import ComplementNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, classification_report
```

```python
from google.colab import drive
drive.mount('/content/drive/')

data = pd.read_csv("/content/drive/MyDrive/NLP Datasets/hatespeech.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
## Counting the null objects
data.isnull().sum()
```

```python
# Checking for duplicated data
data.duplicated().sum()

# Removing duplicates
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# Lowercasing the text
data['tweet'] = data['tweet'].str.lower()
data['tweet'].head()
```

```python
# Removing "Subject: ", as it is the email template, not the email content
data['tweet'] = data['tweet'].str.replace('@user ', '')
data['tweet'] = data['tweet'].str.replace('@', '')
data['tweet'] = data['tweet'].str.replace('#', '')
data['tweet'].head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['tweet'], data['Toxicity'], test_size=0.1, shuffle=True, random_state=1)
```

```python
# Bag-of-Words
vectorizer = CountVectorizer()
vectorized = vectorizer.fit(data['tweet'])

train_vectorized_X = vectorizer.transform(X_train).toarray()
test_vectorized_X = vectorizer.transform(X_test).toarray()
```

```python
naive_bayes = ComplementNB()
naive_bayes.fit(train_vectorized_X, y_train)
predictions = naive_bayes.predict(test_vectorized_X)
```

```python
print(classification_report(y_test, predictions,  digits=4))
```

```python
print(confusion_matrix(y_test, predictions))
```
